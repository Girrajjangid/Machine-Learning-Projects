latitude == Latitude of the location at which data was collected.
longitude == Longitude of the location at which data was collected.
humidity == Ambient relative humidity value obtained from the temperature and humidity sensor. Unit: %
temperature == Ambient temperature value obtained from the temperature and humidity sensor. Unit: Degree Celcius
heat_index == Heat index value obtained from temperature and humidity sensor. It can be derived from Temperature and Humidity values. Unit: No units

uv == UV index obtained from the light sensors. UV index is a linear scale which indicates the severity of the UV exposure. The lowest value is 0 and the highest value is 10. Unit: No units

luminence == Ambient light intensity. Unit: Lux

ir == Infrared intensity. Unit: Lux


Accelerometer and Gyroscope - It a 6 degree of freedom(DoF) accelerometer device (MPU6050)

accx == Raw acceleration in X direction

accy == Raw acceleration in Y direction

accz == Raw acceleration in Z direction

acctemp == Temperature reading obtained from MPU6050

gyrox == Raw gyroscope orientation in X drection

gyroy == Raw gyroscope orientation in Y drection

gyroz == Raw gyroscope orientation in Z direction

so2_tmp == Temperature reading obtained from SPEC SO2 sensor

https://www.spec-sensors.com/wp-content/uploads/2016/10/ULPSM-SO2-968-006.pdf

so2_gas

Raw voltage values from SO2 sensor. Unit: mV

so2_conc

SO2 concentration. Unit: ppm. Note: For details on the conversion from raw voltage to SO2 concentration, please refer to the data sheet of SPEC SO2 sensor

nh3

Ammonia concentration as obtained by Multichannel gas sensor. Unit: ppm

http://wiki.seeedstudio.com/Grove-Multichannel_Gas_Sensor/

co

Carbon Monoxide concentraion as obtained by Multichannel gas sensor. Unit: ppm

no2

Nitrogen di-oxide concentraion as obtained by Multichannel gas sensor. Unit: ppm

c3h8

Propane concentraion as obtained by Multichannel gas sensor. Unit: ppm

c4h10

Iso Butane concentraion as obtained by Multichannel gas sensor. Unit: ppm

ch4

Methane concentraion as obtained by Multichannel gas sensor. Unit: ppm

h2

Hydrogen concentraion as obtained by Multichannel gas sensor. Unit: ppm

c2h5o4

Ethanol concentraion as obtained by Multichannel gas sensor. Unit: ppm

pm01

PM 01 concentration. Unit: ug/cc

 
pm25

PM 01 concentration. Unit: ug/cc

 
pm10

PM 01 concentration. Unit: ug/cc

 
pressure

Differential Pressure (Pa) at the sensor.

This sensor(MPXV7002 series) measures the relative air speed at the device using differential pressure. https://www.nxp.com/docs/en/data-sheet/MPXV7002.pdf

draft

Draft

air_temp

Air Temperature(Deg C)
